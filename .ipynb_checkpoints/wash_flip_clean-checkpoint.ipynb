```python
import pandas as pd
import numpy as np
from glob import glob
import re
```

```python
wash_flip_files = sorted(glob('DataCopy/wash_flip_*.csv'))
wash_flip_files
```

```python
wash_flip = pd.concat((pd.read_csv(file).assign(filename=file)
          for file in wash_flip_files), ignore_index=True)
```

```python
wash_flip
```

```python
wash_flip.drop(['add_feat','body_feat','conv_feat','dimensions','gen_spec', 'highlights','prod_desc','wash_mod','filename','pow_feat'], axis=1, inplace=True)
```

```python
wash_flip.head()
```

```python
wash_flip = wash_flip[['product_name','brand', 'product_price', 'product_rating', 'no_ratings']]
```

```python
wash_flip.dtypes
```

```python
wash_flip['weighted_rating'] = wash_flip['no_ratings'] * wash_flip['product_rating'] * 0.05
```

```python
wash_flip.head(30)
```

```python
from numpy import nan
wash_flip = wash_flip[wash_flip['bins'].notna()]
wash_flip.reset_index(drop=True, inplace=True)
wash_flip
```

```python
cut_labels = ['<15k', '<20k', '<25k', '<30k', '<35k', '>35k']
cut_bins = [0, 15000, 20000, 25000, 30000, 35000, 200000]
wash_flip['bins'] = pd.cut(wash_flip['product_price'], bins=cut_bins, labels=cut_labels)
wash_flip.head(30)
```

```python
wash_flip = wash_flip.sort_values(['bins', 'weighted_rating'], ascending=False)
wash_flip.reset_index(drop=True, inplace=True)
wash_flip.head(30)
```

```python
wash_flip.to_csv('wash_flip_bins.csv', sep=',', index=False)
```
