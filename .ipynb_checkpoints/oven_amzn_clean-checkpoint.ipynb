```python
import pandas as pd
import numpy as np
from glob import glob
import re
```

```python
oven_amzn_files = sorted(glob('DataCopy/oven_amzn_*.csv'))
oven_amzn_files
```

```python
oven_amzn = pd.concat((pd.read_csv(file).assign(filename=file)
          for file in oven_amzn_files), ignore_index=True)
```

```python
oven_amzn
```

```python
oven_amzn.drop(['highlights','prod_desc','tech_details','filename'], axis=1, inplace=True)
```

```python
oven_amzn
```

```python
oven_amzn = oven_amzn[['product_name', 'product_price', 'product_rating', 'no_ratings']]
oven_amzn
```

```python
oven_amzn.dtypes
```

## Change datatype
Can't multiply float and object, so first cconvert object to float
Now before doing that, remove commas from the numbers

```python
oven_amzn["no_ratings"] = oven_amzn["no_ratings"].astype(float)
oven_amzn["product_price"] = oven_amzn["product_price"].astype(float)
```

```python
oven_amzn['weighted_rating'] = oven_amzn['no_ratings'] * oven_amzn['product_rating'] * 0.05
```

```python
oven_amzn
```

```python
cut_labels = ['<10k', '<15k', '<20k', '>21k']
cut_bins = [0, 10000, 15000, 20000, 100000]
oven_amzn['bins'] = pd.cut(oven_amzn['product_price'], bins=cut_bins, labels=cut_labels)
```

```python
from numpy import nan
oven_amzn = oven_amzn[oven_amzn['bins'].notna()]
oven_amzn.reset_index(drop=True, inplace=True)
oven_amzn
```

```python
oven_amzn.groupby('weighted_rating', axis=1)
oven_amzn
```
